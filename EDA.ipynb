```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset with dtype specification to prevent mixed types
df = pd.read_csv(
    r"C:\Users\The Best\Desktop\Projects\Microsoft Cyber\Raw data\GUIDE_Test.csv",
    dtype={'Column1': str, 'Column2': float},  # Specify the columns with mixed types, replace with actual column names
    low_memory=False
)

# Step 2: Inspect the dataset
print(df.head())           # Display the first few rows of the dataset
print(df.info())           # Check for the data types and non-null values
print(df.columns)          # Check for the column names
print(df.isnull().sum())   # Check for missing values

# Step 3: Check descriptive statistics for numeric columns
print(df.describe())

# Step 4: Define the target variables based on actual column names
# Replace these with actual column names that represent your target variables
target_variables = ['IncidentGrade', 'SuspicionLevel', 'LastVerdict']  # Update based on df.columns output

# Step 5: Plot the distribution of each target variable
plt.figure(figsize=(18, 5))

for i, target in enumerate(target_variables, 1):
    if target in df.columns:
        plt.subplot(1, 3, i)
        sns.countplot(x=target, data=df)
        plt.title(f'Distribution of {target}')
        plt.tight_layout()
    else:
        print(f"Column '{target}' not found in the dataset")

plt.show()

# Step 6: Plot Correlation Heatmap for numeric features only
numeric_df = df.select_dtypes(include='number')

if not numeric_df.empty:
    plt.figure(figsize=(10, 8))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.show()
else:
    print("No numeric data available for correlation heatmap.")

```
